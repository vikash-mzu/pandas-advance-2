```python
Q1. Code to Print the Data Present in the Second Row of the DataFrame df

import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Printing the second row
print(df.iloc[1])



Q2. Difference Between loc and iloc in pandas.DataFrame
loc: Accesses data using labels. It includes the end value in slicing.
iloc: Accesses data using integer-based indices. It does not include the end value in slicing.
For example:


# Using loc
print(df.loc[1])  # Access row with index label 1

# Using iloc
print(df.iloc[1])  # Access the second row (index-based)


    
Q3. Reindex the DataFrame and Explain Differences Between loc and iloc

# Reindex the DataFrame
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# loc and iloc outputs
print(new_df.loc[2])
print(new_df.iloc[2])
Explanation:

new_df.loc[2] accesses the row labeled with the index 2 (original row).
new_df.iloc[2] accesses the third row based on the new index position.
Difference: loc accesses by label, while iloc accesses by position. If the DataFrame is reindexed, loc will fetch data based on the new label, and iloc will fetch based on positional indexing.

Q4. Statistical Measurements for df1

import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# (i) Mean of each column
print(df1.mean())

# (ii) Standard deviation of column_2
print(df1['column_2'].std())



Q5. Replace Data in Second Row of column_2 and Calculate Mean

# Replacing the value in the second row of column_2 with a string
df1.loc[2, 'column_2'] = 'string_value'

# Now trying to calculate the mean will cause an error
try:
    print(df1['column_2'].mean())
except Exception as e:
    print(f"Error: {e}")
Explanation: The error occurs because pandas cannot calculate the mean of a column that contains non-numeric data (e.g., a string).

Q6. Understanding Pandas Window Functions
Window functions allow you to perform operations like rolling averages, cumulative sums, etc., over a sliding window of data.

Types of window functions:

Rolling window: Performs calculations over a moving window of a fixed size.
Expanding window: Expands the window to include all previous data points.
Example:


df['RollingMean'] = df['column_name'].rolling(window=3).mean()




Q7. Code to Print the Current Month and Year


import pandas as pd

# Get current date
current_date = pd.Timestamp.now()

# Extract month and year
print(current_date.strftime('%B %Y'))




Q8. Calculate Difference Between Two Dates

import pandas as pd

def calculate_time_difference(date1, date2):
    date1 = pd.to_datetime(date1)
    date2 = pd.to_datetime(date2)
    
    # Difference in timedelta
    delta = date2 - date1
    
    days = delta.days
    seconds = delta.seconds
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    
    print(f"Difference: {days} days, {hours} hours, {minutes} minutes")

# Taking input
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

calculate_time_difference(date1, date2)



Q9. Convert a Column to Categorical and Sort the Data

import pandas as pd

def convert_column_to_category(file_path, column_name, category_order):
    df = pd.read_csv(file_path)
    
    # Convert the specified column to categorical
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    
    # Sort the data
    df_sorted = df.sort_values(by=column_name)
    
    print(df_sorted)

# Inputs
file_path = input("Enter the file path: ")
column_name = input("Enter the column name: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

convert_column_to_category(file_path, column_name, category_order)




Q10. Visualize Sales Data Using a Stacked Bar Chart

import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales(file_path):
    df = pd.read_csv(file_path)
    
    # Create a pivot table for the stacked bar chart
    df_pivot = df.pivot_table(index='Date', columns='Product_Category', values='Sales', aggfunc='sum')
    
    # Plot the stacked bar chart
    df_pivot.plot(kind='bar', stacked=True)
    
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.title('Sales by Product Category Over Time')
    plt.show()

# Input file path
file_path = input("Enter the file path: ")
visualize_sales(file_path)

import pandas as pd

def calculate_statistics(file_path):
    df = pd.read_csv(file_path)
    
    mean = df['Test Score'].mean()
    median = df['Test Score'].median()
    mode = df['Test Score'].mode().values
    
    # Display the results in a table format
    print(f"+-----------+--------+")
    print(f"| Statistic | Value  |")
    print(f"+-----------+--------+")
    print(f"| Mean      | {mean:.1f}    |")
    print(f"| Median    | {median:.1f}  |")
    print(f"| Mode      | {', '.join(map(str, mode))} |")
    print(f"+-----------+--------+")

# Input file path
file_path = input("Enter the file path of the CSV file: ")
calculate_statistics(file_path)
```
